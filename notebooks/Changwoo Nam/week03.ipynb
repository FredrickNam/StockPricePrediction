## 초론 + 서론 정독
* LSTM 기법을 노이즈가 강한 주식 데이터에 적용하여 주가 방향성 예측 성능 향상
    * 금융 시계열 데이터는 높은 노이즈와 시장 효율성으로 인해 예측이 매우 어려움
    * LSTM은 시퀀스 학습(Sequence learning)에 매우 뛰어난 최첨단 기법으로, 금융 시계열 데이터에 본질적으로 적합함에도 불구하고 그동안 널리 적용되지 않았음
    * 해당 논문은 1992년부터 2015년까지 S&P 500 구성 종목의 방향성 움직임을 예측하기 위해 LSTM 네트워크를 배포
    * 실험 결과, LSTM 모델은 거래 비용 공제 전 일일 수익률 0.46%를 기록하며, 기존의 메모리가 없는 모델들인 랜덤 포레스트(RAF), 심층 신경망(DNN), 로지스틱 회귀(LOG)의 예측 성능을 능가

* 모델의 블랙박스적 측면을 타파하고 실제 선정되는 주식들의 특징 파악
    * 인공 신경망의 고질적인 문제인 블랙박스(Black-box) 특성을 조명하여 모델의 수익성 원천을 파악하고자 함
    * LSTM이 거래를 위해 선택한 주식들은 공통적으로 높은 변동성(high volatility), 평균 이하의 모멘텀, 그리고 단기 반전(short-term reversal)의 특징을 보인다는 것을 밝혀냄

* 데이터 전처리, 모델링, 모델 학습, 주가 데이터에 적용
    * 규모가 크고 생존 편향(survivor bias)이 없는 주식 유니버스를 대상으로 LSTM을 적용하기 위해 포괄적인 가이드를 제공
    * 데이터 전처리(data preprocessing)부터 모델의 개발(development), 학습(training), 배포(deployment)까지의 구체적인 과정이 포함

* 모델이 선정하는 규칙의 핵심 파악 및 전략 구축
    * LSTM 모델이 패턴을 식별하는 방식을 바탕으로 단순화된 규칙 기반(rules-based) 단기 반전 거래 전략을 공식화
    * 단기적으로 극단적인 하락을 보인 종목(losers)을 매수하고 상승을 보인 종목(winners)을 공매도하는 이 전략은 거래 비용 전 일일 0.23%의 수익을 창출
    * 회귀 분석을 통해 LSTM 모델이 도출한 수익은 기존 벤치마크 모델들과 비교했을 때 일반적인 시스템적 위험 요인(systematic risk)에 대한 노출도가 낮다는 것을 확인
